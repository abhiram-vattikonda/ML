```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("tips.csv")

data.head()
```

```python
t_data = data.copy()
t_data['percent_tip'] = round(t_data['tip']*100 / t_data['total_bill'], 1)
per_count = t_data['percent_tip'].value_counts()
t_data.head()
per_count
```

```python
# What are the min, max, median, quartiles or tip percentage
# - The min is around 3, max is 29. There are a few outliers in the data.
# - Most of the tips are in the range of 12 to 20 percent

sns.boxplot(y = 'percent_tip', data = t_data)
plt.ylabel('Tip Percentage')
plt.title('Distribution of Tip Percentage')
plt.show()
```

```python
# What is teh distribution of total sales??
# - Most common bill amount is around 17-18

sns.kdeplot(t_data['total_bill'])
plt.show()
```

```python
# On which day was the restaurent visited most
# - Sunday and Saturday had the most visits

sns.countplot(x = 'day', data = t_data)
plt.title('Count of days')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.show()
```

```python
# What percent of people were Smokers?
# - around 40% of the people are smokers
smoker_counts = t_data['smoker'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Smoker Category')
plt.show()
```

```python
# What is the ration of genders that visited the restaurent?
# - around 35% are female and around 65% are male
sex_counts = t_data['sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Sex Category')
plt.show()
```
